```python
# check integrity of v1.0.3 extraction metadata

merged = df_extraction_registry.merge(
    df_base_registry[["output_sha", "software_version", "pipeline_version", "doc_doi", "dependencies"]],
    on="output_sha",
    how="left"
)

v103 = merged[merged["software_version"] == "v1.0.3"]

print("=== v1.0.3 extraction overview ===")
print(v103.groupby(["stage", "section_type"]).size())

print("\n=== missing doc_doi ===")
print(v103["doc_doi"].isna().sum())

print("\n=== missing dependencies ===")
print(v103["dependencies"].isna().sum())

print("\n=== missing file_path ===")
print(v103["file_path"].isna().sum())

print("\n=== files actually on disk ===")
missing_files = []
for _, row in v103.iterrows():
    if pd.notna(row["file_path"]) and not Path(row["file_path"]).exists():
        missing_files.append(row["file_path"])

print(f"Files missing from disk: {len(missing_files)}")
if missing_files:
    print("Sample missing files:")
    for f in missing_files[:5]:
        print(f"  {f}")

print("\n=== duplicate output_sha in extraction registry ===")
print(v103.duplicated(subset=["output_sha"]).sum())

print("\n=== dependency SHAs pointing to existing base registry entries ===")
v103_deps = set(v103["dependencies"].dropna())
base_shas = set(df_base_registry["output_sha"])
missing_deps = v103_deps - base_shas
print(f"Dependencies not found in base registry: {len(missing_deps)}")
```

```python
missing_dep_rows = v103[v103["dependencies"].isin(missing_deps)]
print(missing_dep_rows[["doc_doi", "section_type", "stage", "dependencies"]])
```

```python
# get the affected dois and their dependency shas
affected = v103[v103["dependencies"].isin(missing_deps)][["doc_doi", "dependencies"]].drop_duplicates(subset=["dependencies"])

for _, row in affected.iterrows():
    doc_doi = row["doc_doi"]
    expected_sha = row["dependencies"]
    folder_doi = doc_doi.replace("/", "_")
    
    # find the txt file in v1.0.0
    v100_folder = RECORDS_DIR / "v1.0.0" / "pipeline_version" / "v1.0.0" / folder_doi
    
    txt_files = [f for f in v100_folder.glob("*.txt")
                 if f.name not in ("DAS_section.txt", "DAS_section_cleaned.txt",
                                   "CAS_section.txt", "CAS_section_cleaned.txt")]
    
    if not txt_files:
        print(f"No txt file found on disk for {doc_doi}, cannot reconstruct")
        continue
    
    txt_path = txt_files[0]
    actual_sha = compute_hashes(txt_path)
    
    if actual_sha != expected_sha:
        print(f"SHA mismatch for {doc_doi}: expected {expected_sha}, got {actual_sha}")
        continue
    
    # sha matches, reconstruct the base registry entry
    base_row = {
        "output_sha": actual_sha,
        "doc_doi": doc_doi,
        "output_type": "raw file",
        "pipeline_version": "v1.0.0",
        "software_version": "v1.0.0",
        "creation_date": datetime.now(),
        "dependencies": None  # txt files have no dependency, they come from PDFs
    }
    
    df_base_registry = pd.concat([df_base_registry, pd.DataFrame([base_row])], ignore_index=True)
    print(f"Reconstructed base entry for {doc_doi}")

save_registry(df_base_registry, BASE_REGISTRY)

# verify
v103_deps = set(v103["dependencies"].dropna())
base_shas = set(df_base_registry["output_sha"])
still_missing = v103_deps - base_shas
print(f"\nRemaining missing dependencies: {len(still_missing)}")
```
